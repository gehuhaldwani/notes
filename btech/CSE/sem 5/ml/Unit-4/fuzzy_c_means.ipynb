```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(42)
x = np.random.normal(0, 0.5, 100)
y = np.random.normal(0, 0.5, 100)
data = np.vstack((x, y))

# Number of clusters
c = 3

# Apply Fuzzy C-Means
cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(data, c, m=2, error=0.005, maxiter=1000)

# Get the cluster membership for each data point
cluster_membership = np.argmax(u, axis=0)

# Plot the data points and cluster centers
plt.scatter(x, y, c=cluster_membership, cmap='viridis', label='Data Points')
plt.scatter(cntr[:, 0], cntr[:, 1], c='red', marker='x', s=100, label='Cluster Centers')
plt.title("Fuzzy C-Means Clustering")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```
